```python
# import module
from datetime import time

#Creating a time object
#time(hour, minute, second, microsecond) 
t = time(14, 42, 5, 566) 
print ("time =",t)
```

```python
# import module and class
from datetime import datetime

dt = datetime (2017,1,28,23,55,59,1023)
print("year =", dt.year)
print("month =", dt.month)
print("day =", dt.day)
print("hour =", dt.hour)
print("minute =", dt.minute)
print("seconds =", dt.second)
print("micro seconds =", dt.microsecond)
```

```python
print("datetime =",dt)
```

```python
# current date and time 
now = datetime.now() 
print('Current time', now)
```

```python
#printing current year, month and day
from datetime import date
#date object of today's date
today = date.today()

print("Current year:", today.year)
print("Current month:", today.month)
print("Current day:", today.day)
```

```python
import pandas as pd

dr = pd.date_range(start='2022-01-01', end='2022-03-01')
df = pd.DataFrame()
df['Date'] = dr
df['Week_Day'] = pd.to_datetime(dr).weekday
df['Day_name'] = pd.to_datetime(dr).strftime("%A")
df['Month_name'] = pd.to_datetime(dr).strftime("%B")
df
```
